### Features explanation

The meaning of each parameter in the given variable description:

1. **Patient ID:** This is a unique identifier for each patient, allowing you to distinguish between different individuals in your dataset.

2. **Persistency_Flag:** This flag indicates whether a patient was persistent with their therapy or not. It likely distinguishes between patients who continued their prescribed treatment and those who did not.

3. **Age:** Age of the patient during their therapy. This parameter records the age of each patient at the time they underwent therapy.

4. **Race:** Race of the patient. It records the racial background or ethnicity of each patient.

5. **Region:** The region of the patient. This parameter might specify the geographical region or location where each patient is from.

6. **Ethnicity:** Ethnicity of the patient. It records the ethnic background of each patient.

7. **Gender:** Gender of the patient, specifying whether the patient is male or female.

8. **IDN Indicator:** This is a flag that indicates whether patients are mapped to Integrated Delivery Networks (IDNs). IDNs are networks of healthcare providers.

9. **NTM - Physician Specialty:** This parameter records the specialty of the healthcare professional (HCP) who prescribed the NTM (New Therapy Medication) prescription.

10. **NTM - T-Score:** T Score of the patient at the time of the NTM Rx. The T score is a measure used in bone density scans (DXA scans) to assess the risk of osteoporosis.

11. **Change in T Score:** This parameter indicates the change in T score before and after receiving therapy. It categorizes the change as "Worsened," "Remained Same," "Improved," or "Unknown."

12. **NTM - Risk Segment:** Risk Segment of the patient at the time of the NTM Rx. It likely refers to the patient's risk category or level for a specific condition or disease.

13. **Change in Risk Segment:** Similar to the change in T score, this parameter indicates the change in the patient's risk segment before and after receiving therapy.

14. **NTM - Multiple Risk Factors:** This flag indicates if the patient falls under multiple risk categories at the time of the NTM Rx.

15. **NTM - Dexa Scan Frequency:** The number of DEXA scans taken by the patient before the first NTM Rx date. DEXA scans are used to assess bone density.

16. **NTM - Dexa Scan Recency:** This flag indicates the presence of a DEXA scan before the NTM Rx, considering a specific time frame.

17. **Dexa During Therapy:** This flag indicates if the patient had a DEXA scan during their first continuous therapy.

18. **NTM - Fragility Fracture Recency:** This flag indicates if the patient had a recent fragility fracture within a specific time frame.

19. **Fragility Fracture During Therapy:** Similar to Dexa During Therapy, this flag indicates if the patient had a fragility fracture during their first continuous therapy.

20. **NTM - Glucocorticoid Recency:** This flag indicates whether the patient used Glucocorticoids with a certain strength within the one-year look-back period from the first NTM Rx.

21. **Glucocorticoid Usage During Therapy:** This flag indicates if the patient used Glucocorticoids during their first continuous therapy.

22. **NTM - Injectable Experience:** This flag indicates any usage of injectable drugs in the recent 12 months before the NTM OP Rx.

23. **NTM - Risk Factors:** This parameter relates to the risk factors that the patient is falling into, with different criteria for chronic and non-chronic risk factors.

24. **NTM - Comorbidity:** Comorbidities are divided into two categories: acute and chronic, based on ICD codes. This parameter tracks comorbid conditions in patients.

25. **NTM - Concomitancy:** This parameter relates to concomitant drugs recorded prior to starting therapy, within a specified time frame.

26. **Adherence:** Adherence measures how well a patient follows their prescribed therapies. It provides information on whether patients are taking their medications as recommended.

These parameters collectively provide a comprehensive overview of patient information, medical history, and treatment-related data, which can be useful for healthcare analysis and decision-making.

## Deduplication

```python
# ! pip install hashlib

import hashlib

def find_potential_duplicates(dataset, columns_to_compare):
    # Create a dictionary to store hash values for each record
    hash_values = {}

    # Iterate through the dataset and compute hash values for selected columns
    for index, record in dataset.iterrows():
        record_values = tuple(record[column] for column in columns_to_compare)

        # Create a hash object
        hash_object = hashlib.md5()

        # Convert record values to bytes and update hash
        for value in record_values:
            hash_object.update(str(value).encode('utf-8'))

        # Get the hexadecimal representation of the hash
        hash_value = hash_object.hexdigest()

        # Store the hash value in the dictionary
        if hash_value in hash_values:
            hash_values[hash_value].append(index)
        else:
            hash_values[hash_value] = [index]

    # Find and return potential duplicate records based on hash values
    potential_duplicates = [indices for indices in hash_values.values() if len(indices) > 1]

    if not potential_duplicates:
        return pd.DataFrame(columns=dataset.columns)  # Return an empty DataFrame
    else:
        duplicate_records_list = []
        for indices in potential_duplicates:
            duplicate_records = dataset.iloc[indices]
            duplicate_records_list.append(duplicate_records)

        return pd.concat(duplicate_records_list, ignore_index=True)
```

```python
# Dataset 
data_ = df_Healtcare

# Columns to be compared for deduplication
columns_to_compare_ = df_Healtcare.columns

# Form a dataframe with all duplicates
potential_duplicates_clean_mdf = find_potential_duplicates(data_, columns_to_compare_)
print("Potential Duplicate Records:")
potential_duplicates_clean_mdf
```

### Gaussian Naive Bayes classifier

```python
### Gaussian Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold

from sklearn import metrics
from sklearn.metrics import accuracy_score, balanced_accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score


def GaussianNB_model(modeling_data, X_train, y_train, X_test, y_test):
    gnb = GaussianNB() 
    accuracy_train = []
    accuracy_test = []
    balanced_accuracy = []
    precision_0 = []
    precision_1 = []
    recall_0 = []
    recall_1 = []
    f1_score = []
    f1_micro = []
    f1_weighted = []
    roc_auc = []
    kf_scores = []
    for i in range(0,len(X_train)):
        # Fit model
        GNB = gnb.fit(X_train[i], y_train[i])
        y_pred = GNB.predict(X_test[i])
        accuracy_test_ = metrics.accuracy_score(y_test[i], y_pred)
        accuracy_train_ = metrics.accuracy_score(y_train[i], GNB.predict(X_train[i]))
        #Performance Metrics and Evaluation
        balanced_accuracy_ = balanced_accuracy_score(y_test[i], y_pred)
        precision_0_ = metrics.precision_score(y_test[i], y_pred, pos_label=0)
        recall_0_ = metrics.recall_score(y_test[i], y_pred, pos_label=0)
        precision_1_ = metrics.precision_score(y_test[i], y_pred, pos_label=1)
        recall_1_ = metrics.recall_score(y_test[i], y_pred, pos_label=1)
        f1_score_ = metrics.f1_score(y_test[i], y_pred, average=None)
        f1_micro_ = metrics.f1_score(y_test[i], y_pred, average='micro')
        f1_weighted_ = metrics.f1_score(y_test[i], y_pred, average='weighted')
        #ROC curve
        fpr, tpr, threshold = metrics.roc_curve(y_test[i], y_pred) 
        roc_auc_ = metrics.auc(fpr, tpr)
        # Append all
        accuracy_train.append(accuracy_train_)
        accuracy_test.append(accuracy_test_)
        balanced_accuracy.append(balanced_accuracy_)
        precision_0.append(precision_0_)
        precision_1.append(precision_1_)
        recall_0.append(recall_0_)
        recall_1.append(recall_1_)
        f1_score.append(f1_score_)
        f1_micro.append(f1_micro_)
        f1_weighted.append(f1_weighted_)
        roc_auc.append(roc_auc_)
        # K-Fold 
        kf = KFold(n_splits=10, shuffle=True, random_state=0)  
        kf_scores_ = cross_val_score(GNB, X_train[i], y_train[i], cv=kf)
        kf_scores.append(kf_scores_.mean())
    Summary = pd.DataFrame()
    Summary['dataset'] = modeling_data
    Summary['Accuracy on test'] = accuracy_test
    Summary['Accuracy on train'] = accuracy_train
    Summary['KFold score'] = kf_scores
    Summary['Balanced Accuracy'] = balanced_accuracy
    Summary['Precision 0'] = precision_0
    Summary['Precision 1'] = precision_1
    Summary['Recall 0'] = recall_0
    Summary['Recall 1'] = recall_1
    Summary['F1 score 0, 1'] = f1_score
    Summary['F1 micro'] = f1_micro
    Summary['F1 weighted'] = f1_weighted
    Summary['ROC AUC'] = roc_auc
    return Summary, gnb
```

```python
from sklearn.ensemble import AdaBoostClassifier

from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold

from sklearn import metrics
from sklearn.metrics import accuracy_score, balanced_accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score

def AdaBoostClassifier_model(estimator, modeling_data, X_train, y_train, X_test, y_test):
    abc = AdaBoostClassifier(base_estimator=estimator, random_state=0)
    accuracy_train = []
    accuracy_test = []
    balanced_accuracy = []
    precision_0 = []
    precision_1 = []
    recall_0 = []
    recall_1 = []
    f1_score = []
    f1_micro = []
    f1_weighted = []
    roc_auc = []
    kf_scores = []
    for i in range(0,len(X_train)):
        # Fit model
        ABC = abc.fit(X_train[i], y_train[i])
        y_pred = ABC.predict(X_test[i])
        accuracy_test_ = metrics.accuracy_score(y_test[i], y_pred)
        accuracy_train_ = metrics.accuracy_score(y_train[i], ABC.predict(X_train[i]))
        #Performance Metrics and Evaluation
        balanced_accuracy_ = balanced_accuracy_score(y_test[i], y_pred)
        precision_0_ = metrics.precision_score(y_test[i], y_pred, pos_label=0)
        recall_0_ = metrics.recall_score(y_test[i], y_pred, pos_label=0)
        precision_1_ = metrics.precision_score(y_test[i], y_pred, pos_label=1)
        recall_1_ = metrics.recall_score(y_test[i], y_pred, pos_label=1)
        f1_score_ = metrics.f1_score(y_test[i], y_pred, average=None)
        f1_micro_ = metrics.f1_score(y_test[i], y_pred, average='micro')
        f1_weighted_ = metrics.f1_score(y_test[i], y_pred, average='weighted')
        #ROC curve
        fpr, tpr, threshold = metrics.roc_curve(y_test[i], y_pred) 
        roc_auc_ = metrics.auc(fpr, tpr)
        # Append all
        accuracy_train.append(accuracy_train_)
        accuracy_test.append(accuracy_test_)
        balanced_accuracy.append(balanced_accuracy_)
        precision_0.append(precision_0_)
        precision_1.append(precision_1_)
        recall_0.append(recall_0_)
        recall_1.append(recall_1_)
        f1_score.append(f1_score_)
        f1_micro.append(f1_micro_)
        f1_weighted.append(f1_weighted_)
        roc_auc.append(roc_auc_)
        # K-Fold 
        kf = KFold(n_splits=10, shuffle=True, random_state=0)  
        kf_scores_ = cross_val_score(ABC, X_train[i], y_train[i], cv=kf)
        kf_scores.append(kf_scores_.mean())
    Summary = pd.DataFrame()
    Summary['dataset'] = modeling_data
    Summary['Accuracy on test'] = accuracy_test
    Summary['Accuracy on train'] = accuracy_train
    Summary['KFold score'] = kf_scores
    Summary['Balanced Accuracy'] = balanced_accuracy
    Summary['Precision 0'] = precision_0
    Summary['Precision 1'] = precision_1
    Summary['Recall 0'] = recall_0
    Summary['Recall 1'] = recall_1
    Summary['F1 score 0, 1'] = f1_score
    Summary['F1 micro'] = f1_micro
    Summary['F1 weighted'] = f1_weighted
    Summary['ROC AUC'] = roc_auc
    return Summary, abc
```
